1. (1 балл) Подготовить мини-корпус (не меньше 4 текстов, примерный общий объём - 3-5 тысяч токенов) с разметкой ключевых слов. 
Предполагается, что вы найдете источник текстов, в котором **уже выделены** ключевые слова.
Укажите источник корпуса и опишите, в каком виде там были представлены ключевые слова.

Как источник текстов я выбрала аннотации статей с Киберленинки по запросу "публичная история", так как статьи целиком слишком объемные. В аннотациях ключевые слова (из тех, которые указаны как ключевые слова всей статьи) выделены жирным шрифтом.

```python
corpus = []

with open('corpus.csv', 'r', encoding='utf-8') as f:
    csv_reader = csv.reader(f)
    for row in csv_reader:
        content = {}
        content['text'] = row[0]
        content['keywords'] = row[1].split(',')
        corpus.append(content)
```

```python
token_count = 0

for item in corpus:
    token_count += len(word_tokenize(item['text']))

print(token_count)
```

3. (2 балла) Применить к этому корпусу 3 метода извлечения ключевых слов на выбор (RAKE, TextRank, tf*idf, OKAPI BM25, ...)

RAKE

```python
morph = MorphAnalyzer()

def normalize_text(text):
    lemmas = []
    for t in simple_word_tokenize(text):
        lemmas.append(morph.parse(t)[0].normal_form)
    return ' '.join(lemmas)
```

```python
stop = stopwords.words('russian')
rake = RAKE.Rake(stop)

for item in corpus:
    rake_keywords = rake.run(normalize_text(item['text']), maxWords=2, minFrequency=2)
    print(rake_keywords)
    if rake_keywords:
        item['rake'] = [k[0] for k in rake_keywords]
    else:
        item['rake'] = []
```

```python
corpus[0]['rake']
```

TextRank

```python
for item in corpus:
    textrank_kw = keywords.keywords(normalize_text(item['text']), language='russian', additional_stopwords=stop, scores=True)
    item['textrank'] = [k[0] for k in textrank_kw]
```

```python
corpus[0]['textrank']
```

KeyBERT

```python
kw_model = KeyBERT('clips/mfaq')
```

```python
for item in corpus:
    bert_kw = kw_model.extract_keywords(normalize_text(item['text']), keyphrase_ngram_range=(1, 2), stop_words=stop)
    print(bert_kw)
    item['bert'] = [k[0] for k in bert_kw]
```

4. (2 балла) Составить морфологические/синтаксические шаблоны для ключевых слов и фраз, выделить соответствующие им подстроки из корпуса (например, именные группы Adj+Noun).
Применить эти фильтры к спискам ключевых слов.

```python
# Возьмем Noun, Adj + Noun, Noun + Noun

str(morph.parse('белый')[0].tag)
```

```python
for item in corpus:
    for instr in ['rake', 'textrank', 'bert']:
        filtered = []
        for kw in item[instr]:
            if len(kw.split()) == 1 and str(morph.parse(kw)[0].tag).startswith('NOUN'):
                filtered.append(kw)
            elif len(kw.split()) == 2:
                first, second = kw.split()
                if str(morph.parse(second)[0].tag).startswith('NOUN') and\
                   (str(morph.parse(first)[0].tag).startswith('NOUN') or str(morph.parse(first)[0].tag).startswith('ADJ')):
                        filtered.append(kw)
        item['filtered_' + instr] = filtered
```

5. (2  балла) Оценить точность, полноту, F-меру выбранных методов относительно эталона:
с учётом морфосинтаксических шаблонов и без них.

```python
# у нас лемматизировано все, кроме эталона. надо это исправить
for item in corpus:
    new_stand = []
    for kw in item['stand']:
        new_stand.append(normalize_text(kw))
    item['stand'] = new_stand
```

```python
corpus[0]['stand']
```

```python
precision = {}
recall = {}
fscore = {}


for instr in ['rake', 'textrank', 'bert', 'filtered_rake', 'filtered_textrank', 'filtered_bert']:
    both = 0
    algo_positive = 0
    stand_positive = 0
    
    for item in corpus:
        both += len(set(item[instr]) & set(item['stand']))
        algo_positive += len(item[instr])
        stand_positive += len(item['stand'])
        
    if len(item[instr]) == 0:
        precision[instr] = 0
    else:
        precision[instr] = both / algo_positive

    recall[instr] = both / stand_positive

    if precision[instr] + recall[instr] == 0:
        fscore[instr] = 0
    else:
        fscore[instr] = 2 * precision[instr] * recall[instr] / (precision[instr] + recall[instr])
```
